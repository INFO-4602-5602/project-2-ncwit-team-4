```python
import pandas as pd
import numpy as np
import csv
import string
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dfDirty = pd.read_csv("https://raw.githubusercontent.com/INFO-4602-5602/project-2-ncwit-team-4/master/Submission/data/Tianlun/race_major.csv",low_memory=False)
dfDirty
```

```python
dfMajor = dfDirty[['Asian','Native Hawaiian/Other Pacific Islander','White','African American','Hispanics','American Indian/Alaska Native']]
dfMajor.reset_index(drop=True)
dfMajor.head()
```

```python
def automajor(major):
    labels = 'Asian', 'Native Hawaiian', 'White', 'African American', 'Hispanics', 'American Indian/Alaska Native'
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink', 'red']
    explode = (0.1, 0, 0, 0, 0, 0)  # explode 1st slice
    #sizes = dfM[dfM['Major Program Name']==major].values.tolist()
    sizes = dfMajor.iloc[major].values.tolist()
    major = dfDirty['Major Program Name'][major]
    
    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
    plt.title("Percentage of each race in %s" %major)
    plt.axis('equal')
```

```python

automajor(10)
```
